```python
import yolov5
import PIL
```

```python
###############################
# Loading and configuring the model
###############################

model = yolov5.load('keremberke/yolov5m-license-plate')


model.conf = 0.25  # NMS confidence threshold
model.iou = 0.45  # NMS IoU threshold
model.agnostic = False  # NMS class-agnostic
model.multi_label = False  # NMS multiple labels per box
model.max_det = 1000  # maximum number of detections per image
```

```python
###############################
# Running model with PIL image
###############################

img_path = '../images/example.jpg'

img = PIL.Image.open(img_path)

results = model(img, size=640)

results.show()
```

```python
###############################
# Running model with image from URL
###############################

img_url = 'https://stage-drupal.car.co.uk/s3fs-public/styles/original_size/public/2019-09/why-are-number-plates-yellow-and-white.jpg?rt1UJUyIi7L2DpS613hFYlI5ng3U4QT3&itok=3SZjXU0B'

results = model(img_url, size=640)
results.show()
```

```python
###############################
# Running model with image from Hub
###############################

import datasets

dataset_name = 'keremberke/license-plate-object-detection'

dataset = datasets.load_dataset(dataset_name, 'mini')

img = dataset['train'][0]['image']

results = model(img, size=640)
results.show()
```
